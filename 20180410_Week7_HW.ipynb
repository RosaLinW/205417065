```python
def factorial2(g): 
    if g > 0 :
        return g*factorial2(g-1) 
    else :
        return 1
```

```python
factorial2(5)
```

```python
def fib(n): # write Fibonacci series up to n
    """Print a Fibonacci series up to n.""" 
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b 
    print()
```

```python
fib(2000)
```

```python
from skimage import io,data

img=io.imread('gommon_3.jpg') 
rows,cols,dims=img.shape 
img2 = img[:,:,0]

for i in range(0,rows): 
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3): #dims=3
            sum = sum + img[i,j,k] 
    img2[i,j] = sum/3 #dims=3

io.imshow(img2) 
io.show()
io.imsave('gommon3.jpg',img2)
```

```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread("gommon_4.jpg")
rows,cols,dims=img.shape

img2 = np.uint8(np.zeros((rows,cols,3))) 
plt.figure(figsize=(8,8)) 
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows): 
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k] 
        for k in range(0,3):
            img2[i,j,k] = np.uint8(sum/3)

plt.subplot(1,2,2) 
plt.title('grey image') 
plt.imshow(img2) 
plt.show()
```

```python
from skimage import io,data
import numpy as np

img = io.imread("gommon_4.jpg")
rows,cols,dims=img.shape
row2 = int(rows/2)
col2 = int(cols/2)
img2 = img

for i in range(0,row2): 
    for j in range(0,cols): 
        for k in range(0,3):
            img2[i,j,k] = img[i,j,k]

for i in range(row2,rows): 
    for j in range(0,cols): 
        for k in range(0,3):
            img2[i,j,k] = img[row2+(-1)*(i-row2),j,k]

io.imshow(img2) 
io.show()
```
